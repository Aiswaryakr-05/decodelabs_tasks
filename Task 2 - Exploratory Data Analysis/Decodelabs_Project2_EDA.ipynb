```python
df['Product'].value_counts().plot(kind='bar')
plt.title('Product Popularity')
plt.xlabel('Product')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.show()
```

```python
df['OrderStatus'].value_counts().plot (kind='bar')
plt.title("Order Status Distribution")
plt.xlabel('OrderStatus')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df['ReferralSource'].value_counts().plot (kind='bar')
plt.title('Referral source')
plt.xlabel('ReferralSource')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
df[['Quantity','UnitPrice','ItemsInCart','TotalPrice']].boxplot()
plt.title('Box Plot of Numerical Variables')
plt.ylabel('Values')
plt.show()
```
